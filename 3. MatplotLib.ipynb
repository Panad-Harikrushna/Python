# pie Chart

```python
education=['B.Tech','M.Tech','B.Sc','M.Sc']
no_of_students=[50,25,15,8]
```

```python
plt.pie(no_of_students,labels=education,autopct='%1.1f%%',explode=[0,0,0.1,0])
plt.show()
```

```python
plt.bar(education,no_of_students)
plt.ylabel("no_of_students")
plt.title("Nareshit_Students")
plt.show()
```

```python
plt.barh(education,no_of_students)
plt.xlabel("no_of_students")
plt.title("NAreshit_students")
plt.show()
```

# Stacked Bar plot

```python
company=['Google','amzn','msft','fb']
revenue=[90,136,89,27]
profit=[40,20,34,12]
plt.bar(company,revenue, label='Revenue')
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
company=['Google','amzn','msft','fb']
revenue=[90,136,89,27]
profit=[40,20,34,12]
xpos=np.arange(len(company))
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.xticks(xpos,company)
plt.ylabel('Revenue(BIn)')
plt.title(("US Technology Stocks"))
plt.legend()
plt.show()
```

- Revenue wise--> Amazon is High
- Profit wise Gooogle is High

```python
#Program for draawing multiple bar charts on one image.
ecommerce=['Snapedeal','Alibaba','Amazon','Flipkart']
Q1_Profit=[45,100,70,40]
Q2_Profit=[40,105,65,45]
Q3_Profit=[42,120,72,50]
Q4_Profit=[34,115,60,48]

#Creating different bar charts on one image using subplot() function.
plt.figure(figsize=(10,10))

# Creating bar chart in first cell of figure having 3 rows, 2 columns,
plt.subplot(3,2,1)
plt.bar(ecommerce,Q1_Profit)
plt.title("Quarter1_Profit")

#Creating bar chart inm second cell
plt.subplot(3,2,2)
plt.bar(ecommerce,Q2_Profit)
plt.title("Quarter2_Profit")

# Creating a bar chart in third cell
plt.subplot(3,2,3)
plt.bar(ecommerce,Q3_Profit)
plt.title("Quarter3_Profit")

# Creating bar chart in the sixth cell
plt.subplot(3,2,6)
plt.bar(ecommerce,Q4_Profit)
plt.title("Quarter4_Profit")


#Display the chart
plt.show()
```
